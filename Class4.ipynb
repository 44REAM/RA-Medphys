## Machine learning vs Deep learning

Machine learning ยังต้องมี domain knowledge ช่วยอยู่ในการเลือกสิ่งที่เราจะส่งเข้าไปให้ model learn ซึ่งในบางครั้ง knowledge ของมนุษย์อาจจะยังมีไม่พอในการทำให้เครื่องสามารถเรียนได้อย่างมีประสิทธิภาพ


### Example

-   Face detection (unstructural data):
    -   ML: มนุษย์จะต้องทำ features extraction เช่นจากรูปหน้าคน เราอาจจะ extract ออกมาว่าตาขนาดเท่าไร ตาห่างกับจมูกเท่าไร แล้วนำค่าพวกนี้ใส่ให้ model learn
    -   DL: สามารถใส่รูปไปเพื่อให้เครื่องเรียนรู้ได้ด้วยตัวเองเลย

-   tabular data เช่นข้อมูลผู้ป่วย เพศ อายุ etc (structural data):
    -   ML: เราอาจจะไม่จำเป็นต้องทำ features extraction เลยก็ได้
    -   DL: ทำได้เหมือนกันกับ ML แต่ตัว model จะ complex มากกว่า

จะเห็นว่าถ้าข้อมูลที่เราต้องการให้เครื่องเรียนรู้มีความซับซ้อนมากขึ้นเช่น รูปภาพหรือข้อมูลที่อยู่ในรูปแบบข้อความ (หนังสือ) DL นั้นมีความสำคัญอย่างมาก
